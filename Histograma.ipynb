## HISTOGRAMA

```python
import cv2 
from matplotlib import pyplot as plt 
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("urso1.jpg")
```

```python
cv2_imshow(img)
```

```python
histr = cv2.calcHist([img],[0],None,[256],[0,256]) 
plt.plot(histr) 
plt.show() 
```

```python
plt.hist(img.ravel(),256,[0,256]) 
plt.show() 
```
